```python
def fasta_to_tsv(fasta_file, txt_file):
    with open(txt_file, 'w') as txt:
        for record in SeqIO.parse(fasta_file, "fasta"):
            identifier = record.id
            sequence = str(record.seq)
            txt.write(f'{identifier}\t{sequence}\n')
```

```python
def count_potential_stop_codons(cds_sequence):
    stop_codons = ['TGA', 'TAG', 'TAA']
    potential_stops = 0

    # Parcourir la sÃ©quence par tranches de 3 nuclÃ©otides (codons)
    for i in range(0, len(cds_sequence) - 2, 3):
        codon = cds_sequence[i:i+3]

        # VÃ©rifier chaque position dans le codon pour un changement potentiel en codon stop
        for j in range(3):
            for nucleotide in ['A', 'T', 'G', 'C']:
                # GÃ©nÃ©rer un nouveau codon avec le nuclÃ©otide modifiÃ©
                mutated_codon = codon[:j] + nucleotide + codon[j+1:]

                # VÃ©rifier si le nouveau codon est un codon stop
                if mutated_codon in stop_codons:
                    potential_stops += 1

    return potential_stops
```

```python

# Appliquer la fonction Ã  la colonne 'CDS' et crÃ©er une nouvelle colonne avec le nombre de codons stop potentiels
df_NM_CDS['Potential_Stop_Count'] = df_NM_CDS['CDS'].apply(count_potential_stop_codons)

df_NM_CDS
```
